```python
from matplotlib import pyplot as plt
import math
```

```python
request_list = [121, 179, 64, 7, 36, 28, 115, 90, 128, 194]
curr_head = 50
max_cylinder = 199
```

```python
def CLOOK(request_list, curr_head, moving_direction):
    temp_list = request_list.copy()
    left_request_request_request = []    
    right_request = []
    x = []
    y = []
    x.append(curr_head)
    if(moving_direction=="left"):
        for i in temp_list:
            if i<curr_head:
                left_request_request_request.append(i)
            else:
                right_request.append(i)
        left_request_request_request.sort(reverse = True)
        for i in left_request_request_request:
            x.append(i)
        right_request.sort(reverse = True)
        for i in right_request:
            x.append(i)
    elif(moving_direction=="right"):
        for i in temp_list:
            if i>curr_head:
                right_request.append(i)
            else:
                left_request_request_request.append(i)
        right_request.sort()
        for i in right_request:
            x.append(i)
        left_request_request_request.sort()
        for i in left_request_request_request:
            x.append(i) 
    
    size = len(x)
    for i in range(0,size):
        y.append(-i)

    plt.figure(figsize=(10, 5))
    plt.xticks(range(0, max_cylinder + 1, 20))
    plt.plot(x,y, color="green", markerfacecolor = 'blue', marker='o', markersize = 5, linewidth = 2, label="CLOOK")
    plt.ylim = (0,size)
    plt.xlim = (0,max_cylinder)
    plt.yticks([])
    plt.title("C-LOOK Scheduling Algorithm")
    plt.show()
```

```python
CLOOK(request_list,curr_head,"right")
```
